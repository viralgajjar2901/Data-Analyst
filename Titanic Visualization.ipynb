```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

```

```python
data=sns.load_dataset('titanic')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.sex.value_counts().plot(kind='pie')
plt.legend()
plt.show()
```

```python
data.shape
```

```python
plt.figure(figsize=(12,8))

plt.scatter(data['age'],data['fare'],color='hotpink')
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')


plt.show()

```

```python
plt.hist(data['age'])
plt.xlabel('Age')
plt.ylabel('Number')

plt.show()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='survived',hue='sex',data=data,palette='Set3')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='survived',hue='pclass',data=data,palette='rainbow')
```

```python
sns.countplot(data['class'])
```

```python
plt.figure(figsize=(10,5))
sns.distplot(data[data['survived']==0]['age'])
sns.distplot(data[data['survived']==1]['age'])
```

```python
sns.jointplot(x=data['survived'],y=data['age'])
```

```python
plt.style.use('fivethirtyeight')
```

```python
sns.relplot(x='age',y='who',data=data,hue='survived')
```

```python
sns.catplot(x='who',y='age',kind='swarm',data=data,hue='survived')
```

```python
sns.boxplot(x=data['age'])
```

```python
sns.catplot(x='',y='survived',kind='box',data=data)
```

```python
sns.scatterplot(x='age',y='alone',data=data,hue='survived')
```
